# Jingju Scores Analysis
This notebook presents some possible uses of the code created for the analysis of the pitch and intervalic structure of the jingju arias contained in the **Jingju Music Scores Dataset** using related manual annotations. For a detailed description of the code, the dataset and the annotions, please refer to the documentation of the [**CMiE2020-JingjuScoreAnalysis**](https://github.com/Rafael-Caro/CMiE2020-JingjuScoresAnalysis) repository. For a thorough description of the functions contained in the code, please refer to their docstrings and the comments within the script.

The main two functions of this code are contained in the `jingjuScoresAnalysis.py` script. So the first step is importing this code. This script draws on auxiliary functions from the `helperFunctions.py` script. So be sure that both scripts are saved in the same folder as this notebook.

```python
import jingjuScoresAnalysis as jsa
```

This code contains two main functions, `pitchHistogram()` and `intervalHistogram()`. Both functions only require two positional arguments, the path to the `line-annotations.csv` file (including the name of the file), which contains annotations for all the lines in the **Jingju Music Scores Dataset**, and the path to the folder where the dataset is stored. So let's create a variable for each of these paths. Make sure that you give the right path according to where you saved the `line-annotations.csv` file and the dataset in your own computer

```python
path2annotations = './annotations/line-annotations.csv'
path2scoresFolder = './JMSD-xml'
```

Now we can call the `pitchHistogram()` and `intervalHistogram()` functions. These functions will perform analyses respectively on pitch and intervals from the lyrics lines in the dataset that match the given musical characteristics. The musical characteristics considered in these two functions, and their available instances, are as follows:

- `roletype`: either `dan` (female character) or `laosheng` (male character)
- `shengqiang` (melodic framework): either `xipi` or `erhuang`
- `banshi` (metrical pattern): either `yuanban` (literally, 'original'), `manban` (slow), or `kuaiban` (fast). In jingju, *kuaiban* is only used in *xipi*.
- `linetype`: jingju lyrics are written in couplets, consisting of an opening line and a closing line. Besides, *erhuang* has two types of opening lines, an extended one, and a compressed one. Therefore, the available instances are:
    - `o1`: extended opening line in *erhuang*
    - `o2`: compressed opening line in *erhuang*
    - `o`: opening line in *xipi*
    - `c`: closing line

You can specify which instances for each musical characteristic you would like to analyze, by inputing them as a list (even if it is one single element). By default, all the instances for all characteristics are considered. So, if no instance for a particular characteristic is given, all of the instances will be analysed. Therefore, if no musical characteristic is specified at all, the analysis is performed on the whole dataset.

Let's first run the `pitchHistogram()` function on the whole dataset.

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder)
```

Both functions have a keyword parameter, `makePlot`, set by default to `False`. If it is given as `True`, a bar chart is plotted with the results.

Let's now run the `intervalHistogram()` function on the whole dataset, setting `makePlot` to `True`.

```python
jsa.intervalHistogram(path2annotations, path2scoresFolder, makePlot=True)
```

Now let's do some more detailed analysis. The pitch range in *dan* and *laosheng* role types should be different in order to express their gender associations. Let's compute pitch histograms for both *dan* and *laosheng*. Remember to give the instances of each musical feature as lists.

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['dan'], makePlot=True)
```

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['laosheng'], makePlot=True)
```

Let's now compare the intervalic structure of the two main melodic frameworks in jingju, *xipi* and *erhuang*.

```python
jsa.intervalHistogram(path2annotations, path2scoresFolder, shengqiang=['xipi'], makePlot=True)
```

```python
jsa.intervalHistogram(path2annotations, path2scoresFolder, shengqiang=['erhuang'], makePlot=True)
```

Have the different *banshi* (metrical patterns) an influence in the pitch structure of *xipi*? Let's compare the three different *banshi* in *laosheng xipi*.

The `pitchHistogram()` function computes the occurrence of pitch in terms of aggregated duration by default. However, by setting the keyword parameter `duration` as `False`, the occurrence is counted just by number of notes.

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['laosheng'], shengqiang=['xipi'], banshi=['yuanban'],
                   duration=False, makePlot=True)
```

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['laosheng'], shengqiang=['xipi'], banshi=['manban'],
                   duration=False, makePlot=True)
```

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['laosheng'], shengqiang=['xipi'], banshi=['kuaiban'],
                   duration=False, makePlot=True)
```

How about the changes in the intervalic structure across *banshi*? Let's compute it for *manban* and *yuanban* in *dan erhuang* (*kuaiban* is not used in *erhuang*).

The `intervalHistogram()` function analyses intervals without considering their direction by default. This can be modified by setting the keyword parameter `directed` to `True`.

Both functions compute the occurrence as percentage by default. This can be also modified by setting the `percentage` parameter to `False`. In this case, the result is given as the absolute count of occurrences.

```python
jsa.intervalHistogram(path2annotations, path2scoresFolder, roletype=['dan'], shengqiang=['erhuang'], banshi=['yuanban'],
                   directed=True, percentage=False, makePlot=True)
```

```python
jsa.intervalHistogram(path2annotations, path2scoresFolder, roletype=['dan'], shengqiang=['erhuang'], banshi=['manban'],
                   directed=True, percentage=False, makePlot=True)
```

Are there significant pitch differences in the two types of opening lines in *erhuang*? Let's compute it for *yuanban*, the 'original' metrical pattern.

The `pitchInterval()` function counts grace notes by default. This can be modified by setting the `gracenotes` keyword parameter to `False`. This might be useful if you are interested in computing results in absolute occurrence of notes.

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['laosheng'], shengqiang=['erhuang'], banshi=['yuanban'],
                   linetype=['o1'], gracenotes=False, percentage=False, duration=False, makePlot=True)
```

```python
jsa.pitchHistogram(path2annotations, path2scoresFolder, roletype=['laosheng'], shengqiang=['erhuang'], banshi=['yuanban'],
                   linetype=['o2'], gracenotes=False, percentage=False, duration=False, makePlot=True)
```
